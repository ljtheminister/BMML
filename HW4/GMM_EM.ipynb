```python
import numpy as np
import pandas as pd
import random
from numpy.linalg import norm
from numpy import exp
from numpy import inf
from numpy import sqrt
from numpy import prod
from numpy import diag
from numpy.linalg import inv
#from scipy.stats import multivariate_normal
from Multivariate_Normal import Multivariate_Normal

class GMM_EM:
    def __init__(self, X, K, pi='Uniform', max_iter=100):
        self.X = X
        self.N, self.p = X.shape
        self.K = K
        self.pi = pi
        if pi == 'Uniform':
            self.pi = np.ones(K)/float(K)
        self.max_iter = max_iter
        self.means = {}
        self.covs = {}
        self.assignments = np.zeros((self.N, self.K))
        self.N_K = {}
        self.converged = False
        # instantiate cluster data structures
        for i in xrange(self.K):
            self.N_K[i] = 0

    def find_nearest_cluster(self, x):
        min_dist = inf
        min_cluster = inf
        for k in xrange(self.K):
            dist = norm(self.means[k]-x)
            if dist < min_dist:
                min_dist = dist
                min_cluster = k
        return min_cluster

    
    def initialize_clusters(self):
        idx = [i for i in xrange(self.N)]
        random.shuffle(idx)
        # randomly initialize K clusters (hard assignment)
        for i in xrange(self.K):
            self.means[i] = self.X[idx[i],:]
        # perform K-means assignment to complete initialization
        for i,x in enumerate(self.X):
            c = self.find_nearest_cluster(x)
            self.assignments[i,c] = 1 # hard assignment

    # M-step
    def M_step(self):
        for k in xrange(self.K):
            # compute N_k for each cluster
            self.N_K[k] = sum(self.assignments[:,k])
            # re-estimate means 
            mean_sum = 0
            for i,x in enumerate(self.X):
                mean_sum += self.assignments[i,k]*x
            # update means
            self.means[k] = mean_sum/float(self.N_K[k])

        for k in xrange(self.K):
            # re-estimate covariances
            cov_sum = np.zeros((self.p,self.p))
            for i,x in enumerate(self.X):
                print np.outer(x-self.means[k], x-self.means[k])
                cov_sum += self.assignments[i,k]*np.outer(x - self.means[k], x - self.means[k])
            # update covariances        
            self.covs[k] = cov_sum/float(self.N_K[k])
            # update mixing proportions
            self.pi[k] = self.N_K[k]/float(self.N)
   
    # E-step
    def E_step(self):
        for i,x in enumerate(self.X):
            row_sum = 0.0
            for k in xrange(self.K):
                mvn = Multivariate_Normal(self.p, self.means[k], self.covs[k])
                responsibility = self.pi[k]*mvn.pdf(x)
                #print "pi: ", self.pi
                #print mvn.pdf(x)
                self.assignments[i,k] = responsibility
                row_sum += responsibility
            self.assignments[i,:] /= float(row_sum)


    # EM Algorithm
    def EM(self):
        # initialize with K-Means
        self.initialize_clusters()
        iter = 0
        while not self.converged and iter < self.max_iter:
            iter += 1
            print 'Iteration: ', iter
            print 'Maximization'
            self.M_step()
            print 'Expectation'
            self.E_step()
```
